TASK-3- Image to Pencil Sketch with Python:

We need to read the image in RBG format and then convert it to a grayscale image. This will turn an image into a classic black and white photo. Then the next thing to do is invert the grayscale image also called negative image, this will be our inverted grayscale image. Inversion can be used to enhance details. Then we can finally create the pencil sketch by mixing the grayscale image with the inverted blurry image. This can be done by dividing the grayscale image by the inverted blurry image. Since images are just arrays, we can easily do this programmatically using the divide function from the cv2 library in Python.

For this task i'm using two images, first one is of a fish and second one is of a tea leaf.

```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
image_size = (300,300)
```

```python
image1 = cv2.imread("/content/drive/MyDrive/Angling Freshwater Fish Netherlands dataset/Abramis brama/0b23d96b-731d-4236-ab1e-11182a282551-340mm.jpg")
image1 = cv2.resize(image1, image_size)
cv2_imshow(image1)
cv2.waitKey(0)
```

```python
image2 = cv2.imread("/content/drive/MyDrive/tea sickness dataset/algal leaf/UNADJUSTEDNONRAW_thumb_2b.jpg")
image2 = cv2.resize(image2,image_size)
cv2_imshow(image2)
cv2.waitKey(0)
```

Converting the image to grayscale

```python
g_image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
cv2_imshow(g_image1)
cv2.waitKey(0)
```

```python
g_image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)
cv2_imshow(g_image2)
cv2.waitKey(0)
```

Inverting the image



```python
inverted_image1 = 255 - g_image1
cv2_imshow(inverted_image1)
cv2.waitKey()
```

```python
inverted_image2 = 255 - g_image2
cv2_imshow(inverted_image2)
cv2.waitKey()
```

Now We will Blur the image using Gausian Function 

```python
blur1 = cv2.GaussianBlur(inverted_image1, (21, 21), 0)
cv2_imshow(blur1)
cv2.waitKey()
```

```python
blur2 = cv2.GaussianBlur(inverted_image2, (21, 21), 0)
cv2_imshow(blur2)
cv2.waitKey()
```

Now we will invert the blurred image to achieve the pencil sketch effect

```python
inverted_blurred1 = 255 - blur1
pencil_sketch1 = cv2.divide(g_image1, inverted_blurred1, scale=256.0)
cv2_imshow(pencil_sketch1)
cv2.waitKey(0)
```

```python
inverted_blurred2 = 255 - blur2
pencil_sketch2 = cv2.divide(g_image2, inverted_blurred2, scale=256.0)
cv2_imshow(pencil_sketch2)
cv2.waitKey(0)
```

Side by Side comparision of original image and processed image

```python
cv2_imshow(image1)
cv2_imshow(pencil_sketch1)
cv2.waitKey(0)
```

```python
cv2_imshow(image2)
cv2_imshow(pencil_sketch2)
cv2.waitKey(0)
```
